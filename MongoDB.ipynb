1. What are the key differences between SQL and NoSQL databases?

Ans - SQL and NoSQL databases differ in their structure, scalability, and usage. SQL databases are relational databases that store data in tables with predefined schemas, while NoSQL databases are non-relational and can store data in various formats such as documents, key-value pairs, graphs, or columns.

Key Differences:

1. Data Structure

- SQL: Data is stored in tables with rows and columns.
- NoSQL: Data is stored in flexible formats like documents, key-value pairs, graphs, or wide-column stores.

2. Schema
- SQL: Uses a fixed schema that must be defined before data insertion.
- NoSQL: Uses a dynamic or flexible schema, allowing changes without significant modifications.

3. Scalability
- SQL: Generally scales vertically by increasing hardware resources.
- NoSQL: Generally scales horizontally by adding more servers.

4. Query Languag

- SQL: Uses Structured Query Language (SQL) for data manipulation.
- NoSQL: Uses database-specific query methods and APIs.

5. Transactions

- SQL: Supports ACID properties, ensuring reliable transactions.
- NoSQL: Often follows BASE principles, prioritizing availability and scalability.

6. Best Use Cases

- SQL: Banking systems, inventory management, and enterprise applications.
- NoSQL: Big data applications, social media platforms, and real-time analytics.

4. How does MongoDB ensure high availability using replication?

Ans - MongoDB ensures high availability through the use of Replica Sets. Here is a breakdown of how they function:

   1. Replica Set Structure

A replica set is a group of mongod instances that maintain the same data set. It typically consists of:

  - Primary Node: The only member that receives all write operations.

  - Secondary Nodes: These replicate the primary's operations to stay synchronized.

  2. Automatic Failover
If the primary node becomes unavailable (e.g., hardware failure or network outage), the remaining members hold an election to choose a new primary. This process ensures that the database remains operational with minimal downtime.

 3. Data RedundancyBy maintaining multiple copies of data on different servers, MongoDB protects the system against a single point of failure.

4. Read Scalability (Optional)While the primary handles writes, clients can configure Read Preferences to send read operations to secondary nodes, distributing the load and improving performance.

6. What is the role of indexes in MongoDB, and how do they improve performance?

Ans - The Role of Indexes

Indexes are special data structures that store a small portion of a collection's data set in a traversable form (typically B-trees). They hold the value of a specific field or set of fields, ordered by the value of the field.

Performance Improvement

  - Avoids Collection Scans: Without an index, MongoDB must perform a collection scan (examine every document in a collection) to find a match. With an index, it can limit the number of documents it has to inspect.
    
  - Faster Sorting: Indexes store data in a sorted order, which allows MongoDB to return sorted results efficiently without needing an in-memory sort.

  - Covered Queries: If an index contains all the fields required by a query, MongoDB can return results directly from the index without even loading the documents from disk.

7. Describe the stages of the MongoDB aggregation pipeline?

Ans- The aggregation pipeline is a framework for data processing where documents pass through a multi-stage pipeline that transforms them into an aggregated result. Common stages include:

    $match: Filters documents to pass only those that match specified conditions (similar to find).
    $group: Groups input documents by a specified identifier and applies accumulator expressions (like sum or average) to the grouped data.
    $project: Reshapes documents by adding, removing, or renaming fields.$sort: Reorders the documents in the pipeline based on specified fields.
    $limit / $skip: Controls the number of documents passed to the next stage.
    $unwind: Deconstructs an array field from the input documents to output a document for each element.
    $lookup: Performs a "left outer join" to another collection in the same database to filter in documents from the "joined" collection.

10. What are the ACID properties in the context of MongoDB transactions?

Ans - ACID refers to a set of properties that guarantee database transactions are processed reliably:

  -  Atomicity: Ensures that all operations within a transaction are completed successfully; if any operation fails, the entire transaction is aborted and rolled back.

  - Consistency: Guarantees that a transaction transforms the database from one valid state to another, maintaining all data rules (e.g., schema validation).

  - Isolation: Ensures that concurrent transactions do not interfere with each other. The effects of an ongoing transaction are invisible to others until it is committed.

  - Durability: Once a transaction is committed, the changes are permanent and will survive even in the event of a system failure (written to the journal).

11. What is the purpose of MongoDB’s explain() function?

Ans - The explain() function is a diagnostic tool used to analyze query performance. Its primary purposes include:

  - Query Execution Plan: It provides details on how MongoDB plans to execute a query (e.g., which indexes are being used).

  - Performance Metrics: It returns statistics such as the number of documents scanned, execution time, and whether a collection scan (COLLSCAN) or index scan (IXSCAN) occurred.

  - Optimization: It helps developers identify slow queries and determine if adding or modifying indexes would improve efficiency.

12. How does MongoDB handle schema validation?

Ans - MongoDB handles schema validation using JSON Schema validation rules applied at the collection level. While MongoDB is fundamentally "schemaless," you can enforce structure during update and insert operations through the following:

 - validator object: When creating or modifying a collection, you define a validator using the $jsonSchema operator.

 - Validation Levels:strict: (Default) Applies validation to all inserts and updates.

 - moderate: Applies validation only to existing documents that already fulfill the validation criteria.

 - Validation Actions:error: (Default) Rejects any document that violates the rules.warn: Allows the operation but logs the violation in the MongoDB log file.

13. What is the difference between a primary and a secondary node in a replica set?

Ans -In a MongoDB replica set, data is stored on multiple servers called nodes.

- Primary Node : Receives all write operations (insert, update, delete) and stores the latest data.

- Secondary Node: Copies data from the primary node and keeps a duplicate copy. It is mainly used for backup and read operations.

Primary Node
- Accepts read and write operations.
- Stores the original data.
- One primary node in a replica set.
- Handles data changes.

Secondary Node
- Usually accepts only read operations.
- Stores a copy of the primary's data.
- Can have multiple secondary nodes.
- Synchronizes changes from the primary node.

14. What security mechanisms does MongoDB provide for data protection?

Ans- MongoDB provides several layers of security to protect data:

 - Authentication : Verifies user identity via mechanisms like SCRAM (default), x.509 certificates, or LDAP/Kerberos integration.
 - Authorization (RBAC) : Uses Role-Based Access Control to restrict user actions. Users are granted specific roles (e.g., readWrite, dbAdmin) rather than full access.
 - Encryption:At Rest : Encrypts data files on disk (available in MongoDB Enterprise and Atlas).
 - In Transit : Uses TLS/SSL to encrypt data moving between the application and the database.
 - Client-Side Field Level Encryption (CSFLE): Encrypts sensitive fields before they even leave the application.
 - Auditing: Tracks system activity and administrative changes for compliance and security monitoring.
 - Network Isolation : Restricts access using firewalls, VPC peering, and IP allowlisting to ensure only trusted sources can connect.

15. Explain the concept of embedded documents and when they should be use?

Ans - Concept

Embedded documents (also known as denormalization) involve nesting related data directly within a single document as sub-documents or arrays. Instead of storing a reference (ID) to a separate collection, all relevant information is stored in one record.

When to Use Them

- One-to-One Relationships : When you have a simple "contains" relationship (e.g., a User has one Address).

- One-to-Few Relationships : When a parent entity has a small, predictable number of child entities (e.g., a Post has a few Tags).

- Read Performance : When you frequently need to retrieve the main entity and its related data at the same time. This avoids multiple database queries or expensive joins.

- Atomicity : When you need to update the main entity and its related data in a single atomic operation.

18. What are the advantages of using MongoDB for horizontal scaling?

Ans - MongoDB uses sharding to achieve horizontal scaling, which offers several key advantages over traditional vertical scaling:

- Cost Efficiency : It allows for scaling out by adding commodity hardware (multiple cheaper servers) rather than upgrading to a single, expensive high-end server.
- Increased Storage Capacity : Data is distributed across multiple machines, allowing the database to grow far beyond the disk limits of a single node.
- Improved Throughput : By distributing read and write operations across many shards, MongoDB reduces the load on any single server, significantly increasing overall performance.
- High Availability : In conjunction with replica sets, sharding ensures that even if one node or shard fails, the database remains operational and data stays accessible.

19. How do MongoDB transactions differ from SQL transactions?

Ans - While both offer ACID (Atomicity, Consistency, Isolation, Durability) guarantees, they differ in scope and design:

- Scope: SQL transactions are designed for a relational model where data is normalized across many tables.
- MongoDB transactions are designed for a document model, where related data is often nested in a single document.
- Need for Transactions: In MongoDB, single-document operations are atomic by default, reducing the need for multi-document transactions. SQL requires transactions for almost any multi-row update.
- Performance: MongoDB transactions generally incur a higher performance cost compared to SQL because MongoDB is optimized for distributed architectures and horizontal scaling.
- Distributed Nature: MongoDB multi-document transactions can span multiple shards and replica sets, whereas standard SQL transactions are typically localized to a single database server.

23. What is MongoDB’s WiredTiger storage engine, and why is it important?

Ans - WiredTiger is the default storage engine for MongoDB (since version 3.2). It is responsible for managing how data is stored on disk and accessed in memory.

Why it is important:

- Document-Level Concurrency : It allows multiple clients to modify different documents in the same collection simultaneously, significantly increasing write throughput.
- Compression : It offers built-in data and index compression (e.g., Snappy or Zlib), reducing storage costs and disk I/O.
- Checkpointing : It uses a "checkpoint" system to ensure data consistency and provide a reliable recovery point in the event of a crash.
- Performance : It efficiently manages a "WiredTiger cache" to keep frequently accessed data in RAM for faster retrieval.

1.  Write a Python script to load the Superstore dataset from a CSV file into MongoDB?

```python
import pandas as pd
from pymongo import MongoClient

# 1. Load the dataset using pandas
# Replace 'superstore.csv' with your actual file path
df = pd.read_csv('superstore.csv', encoding='windows-1252')

# 2. Connect to MongoDB
# Standard local connection; change URI if using MongoDB Atlas
client = MongoClient("mongodb://localhost:27017/")

# 3. Create/Select Database and Collection
db = client["SuperstoreDB"]
collection = db["Sales"]

# 4. Convert DataFrame to a list of dictionaries (JSON-like format)
data_dict = df.to_dict(orient='records')

# 5. Insert data into MongoDB
try:
    collection.insert_many(data_dict)
    print(f"Successfully loaded {len(data_dict)} records into MongoDB.")
except Exception as e:
    print(f"An error occurred: {e}")

# Close the connection
client.close()
```
